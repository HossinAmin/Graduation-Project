```python
import os
import string
import shutil
import random
import cv2
```

```python
ori_ds_dir = "C:\\Users\\10\\Desktop\\Graduation project\\Datasets\\Dataset1\\Dataset cropped\\Training"
my_ds_dir = "C:\\Users\\10\\Desktop\\Graduation project\\Datasets\\Dataset1\\Dataset cropped"

classes = ["A","B","C"]

for f in classes:
    file_list = os.listdir(ori_ds_dir+f"\\{f}")
    img_num = len(file_list)
    train_num = int(img_num*0.8)
    valid_num = int(img_num*0.2)

    for i in range(img_num):
        if i<train_num:
            shutil.copy(ori_ds_dir+f"\\{f}\\"+file_list[i],my_ds_dir+f"\\Training set\\{f}")
        elif i>train_num and i<train_num+valid_num:
            shutil.copy(ori_ds_dir+f"\\{f}\\"+file_list[i],my_ds_dir+f"\\Validation set\\{f}")
```

```python
#padding 
folderDir = "C:\\Users\\10\\Desktop\\Graduation project\\Datasets\\Dataset6\\16 Classes\\Training\\"
for i in os.listdir(folderDir):
   for imgName in os.listdir(folderDir+i+"\\"):
        img = cv2.imread(folderDir+i+"\\"+imgName)
        H,W,_ = img.shape
        if H<=200:
            img = cv2.copyMakeBorder(img, int((200-H)/2), int((200-H)/2), 0, 0, cv2.BORDER_REPLICATE)
        if W<=200:
            img = cv2.copyMakeBorder(img,0,0,int((200-W)/2), int((200-W)/2), cv2.BORDER_REPLICATE)
        
        cv2.imwrite("C:\\Users\\10\\Desktop\\Graduation project\\Datasets\\Dataset6\\16 Classes padded\\"+i+"\\"+imgName,img)
        H,W,_ = img.shape
        print(f"{H},{W}")
```

```python
# finding sizes of ranges
folderDir = "C:\\Users\\10\\Desktop\\Graduation project\\Datasets\\Dataset6\\16 Classes\\Training\\"
for i in os.listdir(folderDir):
    first = True
    H_max = 0
    W_max = 0
    
    for imgName in os.listdir(folderDir+i+"\\"):
        
        img = cv2.imread(folderDir+i+"\\"+imgName)
        H,W,_ = img.shape        
        
        if first:
            H_max = H
            W_max = W
            first = False
    print(f"{i} max h = {H_max} , max w = {W_max}")

        
```
